#Rename Columns

```python
rename_mapping = {
    "SUBDIVISION" : "state_name",
    "YEAR":"crop_year",
    "ANNUAL" : "annual_rainfall_mm",
    "MAM": "summer_rainfall",
    "JJAS" : "kharif_season_rainfall",
    "OND" : "rabi_start_rainfall",
}

rainfall_df.rename(columns = rename_mapping,inplace = True)
print(rainfall_df.head()[['state_name','crop_year','annual_rainfall_mm']].head())
```

```python
rainfall_col = rainfall_df.columns.tolist()
rainfall_col
```

#Handle Missing Values

```python
key_rainfall_cols = ['annual_rainfall_mm','kharif_season_rainfall','rabi_start_rainfall']

#Drop rows where rainfall metrics are missing
rainfall_df_cleaned = rainfall_df.dropna(subset=key_rainfall_cols).copy()
```

```python
print(f"Original Shape: {rainfall_df.shape}")
print(f"Cleaned Shape: {rainfall_df_cleaned.shape}")
```

```python
#Recheck the nulls again
print("Null values check after cleaning: ")
print(rainfall_df_cleaned[key_rainfall_cols].isnull().sum())
```

#Standardize Geographic Names

```python
#Unique Climate State Names
print("Unique Climate State Name (First 10):")
print(rainfall_df_cleaned['state_name'].unique()[:10])
```

```python
rainfall_df_cleaned['state_name'].unique()
```

```python
# STEP 2C: Define and Apply the Mapping to Consolidate IMD Sub-divisions

# Create a mapping dictionary that translates the IMD names (Keys)
# into the desired Agricultural State Names (Values).
state_consolidation_mapping = {
    # Uttar Pradesh
    'East Uttar Pradesh': 'Uttar Pradesh',
    'West Uttar Pradesh': 'Uttar Pradesh',

    # Madhya Pradesh
    'West Madhya Pradesh': 'Madhya Pradesh',
    'East Madhya Pradesh': 'Madhya Pradesh',

    # Maharashtra
    'Konkan & Goa': 'Maharashtra', # Goa data will be included in the Maharashtra analysis
    'Madhya Maharashtra': 'Maharashtra',
    'Matathwada': 'Maharashtra', # Likely a typo for Marathwada
    'Vidarbha': 'Maharashtra',

    # Karnataka
    'Coastal Karnataka': 'Karnataka',
    'North Interior Karnataka': 'Karnataka',
    'South Interior Karnataka': 'Karnataka',

    # Gujarat
    'Gujarat Region': 'Gujarat',
    'Saurashtra & Kutch': 'Gujarat',

    # West Bengal
    'Sub Himalayan West Bengal & Sikkim': 'West Bengal', # Sikkim will be included
    'Gangetic West Bengal': 'West Bengal',

    # Rajasthan
    'West Rajasthan': 'Rajasthan',
    'East Rajasthan': 'Rajasthan',

    # Haryana
    'Haryana Delhi & Chandigarh': 'Haryana', # Delhi/Chandigarh data will be included

    # Andhra Pradesh
    'Coastal Andhra Pradesh': 'Andhra Pradesh',
    'Rayalseema': 'Andhra Pradesh',
    'Telangana': 'Andhra Pradesh', # NOTE: Telangana split from AP in 2014. If your agri data spans years before 2014, this grouping is necessary for time-series consistency.

    # Punjab
    'Punjab': 'Punjab', # No change needed, but kept for clarity

    # Other states not in agri_df (Need to be excluded or handled separately)
    # The remaining regions (e.g., Andaman, Arunachal, Assam, etc.)
    # will be left as their current name, but will naturally be excluded when we merge.
    'Orissa': 'Odisha', # Correcting the common variant name if needed
    'Naga Mani Mizo Tripura': 'North East States', # Consolidate for clarity
}

# Apply the mapping to the 'state_name' column
rainfall_df_cleaned['state_name'] = rainfall_df_cleaned['state_name'].replace(state_consolidation_mapping)

# Final step: Ensure all state names are Title Cased and stripped of whitespace
rainfall_df_cleaned['state_name'] = rainfall_df_cleaned['state_name'].str.strip().str.title()

# Final Verification
print("\nConsolidated Climate State Names (Sample):")
print(rainfall_df_cleaned['state_name'].unique())
```

#Load the agriculture data

```python
agri_df_cleaned = pd.read_csv('/content/drive/MyDrive/My_Projects/Build_For_Bharat/agri_df.csv')
```

```python
##Merge the data
master_df = pd.merge(
    agri_df_cleaned,
    rainfall_df_cleaned,
    on = ['state_name','crop_year'],
    how = 'inner'
)

print("Master Dataset created and merged Successfully")
print(master_df.shape)
```

```python
#Check Numerical Data Types

print("Data Types of key Numerical Columns in Master DF: ")
print(master_df[['production','area','annual_rainfall_mm','kharif_season_rainfall']].dtypes)
```

# Handle Missing Values( NaNs)


```python
#production and area, fill missing values (NaNs) with o
master_df['production'] = master_df['production'].fillna(0)
master_df['area'] = master_df['area'].fillna(0)
```

```python
key_rainfall_col = ['annual_rainfall_mm','kharif_season_rainfall','rabi_start_rainfall']
master_df.dropna(subset=key_rainfall_col,inplace =True)

print(f"\nFinal Master DF shape: {master_df.shape}")
```

```python
# STEP 4C: Save the Final Master Dataset
# Saving to CSV for easy loading by your Q&A system components.

# NOTE: Replace the file path with your desired Google Drive location if needed.
master_df.to_csv('/content/drive/MyDrive/My_Projects/Build_For_Bharat/Combined_Agri_and_Rainfall.csv', index=False)

print("\nFinal Master Dataset saved and ready for Phase 2!")
```
